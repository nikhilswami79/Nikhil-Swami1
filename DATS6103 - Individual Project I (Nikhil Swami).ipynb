```python
#Taking Military Expenditure Colums for the specific years and assigning it to a different Data Frame for Comparision
uf = nf.ix[:,:7]
```

```python
# Data Frame for Finding the variasion in Military Spendings
uf
```

```python
# Getting the Variation in the Military Spending in Fixed Value and Percentage from the Data Set
uf['2010_11'] = uf['ME_2011'] - uf['ME_2010'] 
uf['2011_12'] = uf['ME_2012'] - uf['ME_2011'] 
uf['2012_13'] = uf['ME_2013'] - uf['ME_2012'] 
uf['2013_14'] = uf['ME_2014'] - uf['ME_2013'] 
uf['2014_15'] = uf['ME_2015'] - uf['ME_2014'] 

uf['2010_11_%'] = ((uf['ME_2011'] - uf['ME_2010']) / uf['ME_2010']) * 100
uf['2011_12_%'] = ((uf['ME_2012'] - uf['ME_2011']) / uf['ME_2011']) * 100
uf['2012_13_%'] = ((uf['ME_2013'] - uf['ME_2012']) / uf['ME_2012']) * 100
uf['2013_14_%'] = ((uf['ME_2014'] - uf['ME_2013']) / uf['ME_2013']) * 100
uf['2014_15_%'] = ((uf['ME_2015'] - uf['ME_2014']) / uf['ME_2014']) * 100
```

```python
#DataFrame
uf
```

```python
# Dropping the Columns not Neccessary
uf = uf.drop(['ME_2010','ME_2011','ME_2012','ME_2013','ME_2014','ME_2015'], axis = 1)
```

```python
#DataFrame
uf
```

```python
#Plotting the Variation for the Year 2010-11 in the increasing order for fixed Value
uf[['Country','2010_11']].sort_values(by = '2010_11').plot.bar(x = 'Country', figsize = (18,18), fontsize = '20' , title = 'Fastest Growing Country in terms of Military Expenditure in Fixed Value to the Right for the Year 2010-11')
```
